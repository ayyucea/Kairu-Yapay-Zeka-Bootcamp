## üî¥ ƒ∞leri Seviye

### ‚ùì Soru: "Age" deƒüi≈ükeninin √ßarpƒ±klƒ±k (skewness) ve basƒ±klƒ±ƒüƒ±nƒ± (kurtosis) hesaplayƒ±n ve yorumlayƒ±n.

```python
from scipy.stats import skew, kurtosis
import seaborn as sns

df = sns.load_dataset("titanic")

# Eksik deƒüerleri √ßƒ±karalƒ±m
age_data = df["age"].dropna()

# Skewness ve Kurtosis hesapla
skewness = skew(age_data)
kurt = kurtosis(age_data)

print(f"Skewness (√áarpƒ±klƒ±k): {skewness:.2f}")
print(f"Kurtosis (Basƒ±klƒ±k): {kurt:.2f}")
```

Ya≈ü deƒüi≈ükeni, yakla≈üƒ±k olarak normal bir daƒüƒ±lƒ±ma sahiptir. Hafif saƒüa √ßarpƒ±klƒ±k bulunsa da √ßok ciddi bir sapma ya da u√ß deƒüer yoƒüunluƒüu yoktur.
Basƒ±klƒ±k (kurtosis) deƒüeri ise daƒüƒ±lƒ±mƒ±n normal daƒüƒ±lƒ±ma olduk√ßa yakƒ±n olduƒüunu ve u√ß (aykƒ±rƒ±) deƒüerlerin fazla olmadƒ±ƒüƒ±nƒ± g√∂sterir.
